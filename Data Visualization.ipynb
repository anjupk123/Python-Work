# Data Visualization

```python
#Import library 
import pandas as pd
import numpy as np
```

```python
# read excel file ( the file we cleaned)

Data=pd.read_excel('online_store_customer_data_cleaned.xlsx')

Data.head(5)
```

```python
# Drop unwanted columns

Data.drop(['Unnamed: 0'], axis=1, inplace=True)
Data.head(3)
```

# Line plot

```python
dict_line = {
    'year': [2016, 2017, 2018, 2019, 2020, 2021],
    'price': [200, 250, 260, 220, 280, 300]
}
df_line = pd.DataFrame(dict_line)
```

```python
# use plot() method on the dataframe
df_line.plot('year', 'price');
```

# Bar plot

```python
#For horizontal bar:

Data['Employees_status'].value_counts().plot(kind='bar');
```

```python
#For vertical bar:

Data['Employees_status'].value_counts().plot(kind='barh');
```

# Pie plot

```python
Data['Segment'].value_counts().plot(kind='pie');
```

# Box Plot

```python
#We can plot this by one column or multiple columns. For multiple columns, we need to pass columns name in y variable as a list.

Data.plot(y=['Amount_spent'], kind='box');
```

```python
import matplotlib.pyplot as plt

np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning) 
fig, ax = plt.subplots(figsize=(6,6)) 

Data.boxplot(by ='Employees_status', column =['Amount_spent'],ax=ax, grid = False);
```

# Histogram

```python
Data.plot(y='Age',kind='hist',bins=10);
```

# KDE plot

```python
Data.plot(y='Age',xlim=(0, 100),kind='kde');
```

# Scatterplot

```python
Data.plot(x='Age',y='Amount_spent',kind='scatter');
```
